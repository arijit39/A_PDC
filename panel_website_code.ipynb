```python
data = pd.read_csv('datanew.csv') # data new is the data where cross section id is in front column
```

```python
colname = data.columns
```

```python
noy = 5 # number of years
noce = 6 # number of characters excluded from variable name counting from last
```

```python
a = []
for i in range (1,len(colname),noy):
  l = len(colname[i])-noce
  b = colname[i][0:l:]
  a.append(b)
```

```python
print(a)
```

```python
panel_data = pd.DataFrame()
panel_data['cross_section_id'] = "will be filled"
panel_data['year'] = "will be filled"
c = 0

for j in range (0,len(data),1):

  w = 1

  for i in range (c,c+noy,1):
    panel_data.at[i,"cross_section_id"] = data.iloc[j,0]
    
    if w == 1:
      panel_data.at[i,"year"] = "year_1"
    elif w == 2:
      panel_data.at[i,"year"] = "year_2"
    elif w == 3:
      panel_data.at[i,"year"] = "year_3"
    elif w == 4:
      panel_data.at[i,"year"] = "year_4"
    else :
      panel_data.at[i,"year"] = "year_5"  
    
    num = w 

    for m in range (0,len(a),):
      panel_data.at[i,a[m]] = data.iloc[j,num]
      num = num + noy

    w = w + 1
  c = i + 1

```
